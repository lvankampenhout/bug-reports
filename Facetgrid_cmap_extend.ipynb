```python
import xarray as xr
import numpy as np
import matplotlib
%matplotlib inline
```

```python
xr.__version__
```

```python
airtemps = xr.tutorial.open_dataset('air_temperature')
```

```python
da = airtemps.air
da.dims
```

```python
cmap = 'RdBu_r'
levels = np.arange(240,300+1,10)
```

```python
da[0].plot(cmap=cmap, levels=levels, extend='both')
```

```python
da.sel(time=slice('2013-01-01', '2013-01-02')).plot(cmap=cmap, col='time', col_wrap=4, levels=levels, extend='both')
```
